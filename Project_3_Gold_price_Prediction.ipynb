Importing dependencies



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   # making plots
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
gold_data = pd.read_csv('/content/gld_price_data.csv')  # data processing
```

Correlation
1.  +ve (if 1 tern is incresing then other will also increse)
2.  -ve (if 1 is incresing then other will decrese)

```python
# Drop non-numeric columns
gold_data_numeric = gold_data.select_dtypes(include=['float64', 'int64'])

# Calculate correlation on numeric columns
correlation = gold_data_numeric.corr()
print(correlation)
```

```python
plt.figure(figsize = (8,8))
sns.heatmap(correlation,cbar = True,square = True, fmt='.1f', annot = True,annot_kws = {'size':8},cmap = 'Blues')
```

```python
#corre;ation values of GLD
print(correlation['GLD'])
```

```python
#CHECKING THE DISTRIBUTION OF THE GOLD PRICE
sns.distplot(gold_data['GLD'],color= 'green')
```

Splitting the features and Target

```python
X = gold_data.drop(['Date','GLD'],axis=1)# when u r droping the colunm axis= 1/during thr row axis = 0
Y = gold_data['GLD']
```

```python
print(X)
print(Y)
```

splitting into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

```python
regressor = LinearRegression()
```

```python
# training the model
regressor.fit(X_train,Y_train)

```

Model Evaluation

```python
#prediction on test data
test_data_pretiction = regressor.predict(X_test)
```

```python
print(test_data_pretiction)
```

```python
#R squared error
error_score = metrics.r2_score(Y_test, test_data_pretiction)
print("R squared error : ", error_score)
```

Compare the Actual Value and Predicted Values in plot

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label = 'Actual Value')
plt.plot(test_data_pretiction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
